# Data Visualization / Plotting 

## Task 1 . Plot a histogram of the genre distribution of the movies. How many movies from each genre are there in this list.


```python
# get all generas from the dataset 
genres=movie_df['Genre']    
   
# count unique genres of the dataset 
unique_genres = genres.value_counts()  
```

```python
# set figure size 
plt.figure(figsize=(20,10)) 

# plot histogram 
plt.hist(genres,bins=len(genres.unique()), width=0.9)

# set x and y label 
plt.ylabel('Number of Movies ',fontsize=22)
plt.xlabel('Genres ',fontsize=22)

# set title for the plot 
plt.title("Movie To Genre Distribution: Histogram ",fontsize=30)
plt.xticks(rotation=45,ha= 'right')

plt.show()
```

```python

# extracted data to put on x-axis and y-axis
x_axis_datas= unique_genres.index   
y_axis_datas=unique_genres.values   

plt.figure(figsize=(15,5))

#plot bar chart 
plt.bar(x_axis_datas,y_axis_datas)

# set x-axis and y-axis labels 
plt.ylabel('Number of Movies ' , fontsize=22)
plt.xlabel('Genres ',fontsize=22)
plt.title('Movie to Genre Distribution: Bar Chart ',fontsize=30)
plt.xticks(rotation=45)


plt.show()
```

## Task 2 . Plot a scatterplot in this dataset for the number of movies vs year (Each year, how many movies made to this top 1000 list?)


```python

# extract movies with their release year and count them 

movie_year=movie_df['Released_Year'].value_counts()

# sort movies based on index which is year of release
movie_year=movie_year.sort_index(ascending=True)


plt.figure(figsize=(20,10))  

# set label to the plot 
plt.ylabel('Number of Movies',fontsize=22)
plt.xlabel('Year of Release',fontsize=22)
plt.title("Number of Movies Released Each Year : Scatter Plot",fontsize=30)

# plot scatter diagram 
plt.scatter(movie_year.index,movie_year.values)
plt.xticks(rotation=45,ha= 'right')
plt.tight_layout()


plt.show()
```
